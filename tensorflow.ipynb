### 0. Import Data

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
df = pd.read_csv('Churn.csv')
```

```python
df['Churn'].unique()
```

```python
X = pd.get_dummies(df.drop(['Churn', 'Customer ID'], axis = 1))
y = df['Churn'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
y.unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
X_train = X_train.replace({True: 1, False:0})
X_train.head()
```

```python
X_test = X_test.replace({True: 1, False:0})
X_test.head()
```

```python
y_train.head()
```

### 1. Import Dependencies

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

### 2. Build and Compile model

```python
model = Sequential()
model.add(Dense(units = 32, activation = 'relu', input_dim = len(X_train.columns))) # 'Hidden layer'
model.add(Dense(units = 64, activation = 'relu')) # 'Hidden layer'
model.add(Dense(units = 1, activation = 'sigmoid')) # 'Hidden layer'
```

```python
model.compile(loss = 'binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

### 3. Fit, Predict and Evaluate

```python
model.fit(X_train, y_train, epochs=200, batch_size=32)
```

```python
y_hat = model.predict(X_test)
```

```python
y_hat = [0 if val < 0.5 else 1 for val in y_hat]
```

```python
accuracy_score(y_test, y_hat)
```

### 4. Saving and reloading

```python
model.save('tfmodel')
```

```python
del model
```

```python
model = load_model('tfmodel')
```
